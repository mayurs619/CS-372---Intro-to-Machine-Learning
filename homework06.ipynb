## Part 1: SVM Margins and Loss

### Problem 1.1 (Exam Style)
1. Decision Boundary and Hinge Loss Zero Condition:
        $$
        w^* \cdot x + b^* = 0.
        $$
        The SVM hinge loss for a data point $ (x, y) $ with label $ y \in \{-1, 1\} $ is
        $$
        \ell(w^*, b^*)(x, y) = \max\{0, 1 - y(w^* \cdot x + b^*)\}.
        $$
        The hinge loss is zero when
        $$
        1 - y(w^* \cdot x + b^*) \leq 0 \Rightarrow y(w^* \cdot x + b^*) \geq 1.
        $$
        For any point on the hyperplane $ Z $, we want the loss to be zero, meaning we need
        $$
        y(w^* \cdot x + b^*) = 1.
        $$
   
2. Equation of Hyperplane $ Z $:
        For points with label $ y = 1 $, the hyperplane $ Z $ has the equation
        $$
        w^* \cdot x + b^* = 1.
        $$
        For points with label $ y = -1 $, the hyperplane $ Z $ has the equation
        $$
        w^* \cdot x + b^* = -1.
        $$
        Thus, we can combine both cases by writing
        $$
        w^* \cdot x + b^* = y.
        $$
        Since $ D $ and $ Z $ have equations $ w^* \cdot x + b^* = 0 $ and $ w^* \cdot x + b^* = \pm 1 $, respectively, the two hyperplanes are parallel.


3. Distance Between Parallel Hyperplanes:
        The Euclidean distance between two parallel hyperplanes $ w^* \cdot x + b^* = c_1 $ and $ w^* \cdot x + b^* = c_2 $ is given by
        $$
        \frac{|c_1 - c_2|}{\|w^*\|}.
        $$
        Here, $ c_1 = 0 $ for the decision boundary $ D $, and $ c_2 = \pm 1 $ for the hyperplane $ Z $.
        Therefore, the distance $ d $ between $ D $ and $ Z $ is
        $$
        d = \frac{|0 - (\pm 1)|}{\|w^*\|} = \frac{1}{\|w^*\|}.
        $$
This distance $d = \frac{1}{\|w^*\|}$ confirms that the reference margin $\mu^* = \frac{1}{\|w^*\|}$, as required.

## Part 2: The Representer Theorem for _Softmax_ Classifiers

### Problem 2.1 (Exam Style)
The representer theorem states that for any functional defined on a reproducing kernel Hilbert space (RKHS), the solution can be expressed in terms of the data points. Specifically, if we have a loss function $L(w, b)$ and a regularization term, the optimal parameters $w$ can be expressed as a linear combination of kernel functions evaluated at the training samples.

Functions Definition
- Risk Function $R(a)$:
$$
R(a) = c \|w\|^2
$$

- Regularization Term $S(a_1, \ldots, a_N)$:
$$
S(a_1, \ldots, a_N) = -\frac{1}{N} \sum_{n=1}^{N} \log\left(\frac{e^{y_n(a_n)}}{e^{y_n(a_n) + e^{y_n(-a_n)}}}\right)
$$
where $a_n = b + w \cdot x_n$. This function captures the empirical risk over the training samples.

Assumptions Verification
To check if $L_T(w, b)$ satisfies the assumptions of the representer theorem:
- Continuity: Both $R(a)$ and $S(a_1, \ldots, a_N)$ are continuous functions of $w$ and $b$.
- Convexity/Increasing: $R(a)$ is convex due to the log-sum-exp form, and $S$ is convex since it is a squared norm.
- Finite and Positive: For positive constants $c > 0$, both $R$ and $S$ are finite.

Since both functions satisfy the assumptions, the representer theorem holds for $L_T(w, b)$ when $c > 0$. Therefore, the optimal $w$ can be expressed in terms of the training data.

### Problem 2.2 (Exam Style)
Now, if we set $c = 0$ in the expression for $L_T(w, b)$:
$$
L_T(w, b) = -\frac{1}{N} \sum_{n=1}^{N} \log\left(e^{y_n (b + w \cdot x_n)} + e^{- (b + w \cdot x_n)}\right)
$$
This effectively removes the regularization term.

Functions Definition
- Risk Function $R(a)$:
$$
R(a) = 0
$$

- Regularization Term $S(a_1, \ldots, a_N)$:
$$
S(a_1, \ldots, a_N) = -\frac{1}{N} \sum_{n=1}^{N} \log\left(\frac{e^{y_n(a_n)}}{e^{y_n(a_n) + e^{y_n(-a_n)}}}\right)
$$

Assumptions Verification
- Continuity: $R(a)$ remains continuous.
- Convexity/Increasing: $R(a)$ is not longer convex/strictly increasing.
- Finite and Positive: $R$ is no longer finite nor positive.

Both the Continuity and Increasing conditions of representer point theorem have been violated and thus the representer theorem doesn't hold when $c= 0$.

## Part 3: Linear SVM Classifiers 

```python
from sklearn.datasets import make_classification


def samples(n, d, sep=1):
    xx, yy = make_classification(
        n_samples=n, n_features=d, class_sep=sep, random_state=1
    )
    yy[yy == 0] = -1
    return xx, yy
```

```python
from sklearn.svm import SVC


def linear_svm():
    return SVC(kernel='linear')
```

### Problem 3.1 (Exam Style, except for the programming aspects)

```python
import numpy as np 
training_sample_1_X, training_sample_1_y = samples(1000, 30, 1)
training_sample_20_X, training_sample_20_y = samples(1000,30,20) 

svm_1 = linear_svm().fit(training_sample_1_X, training_sample_1_y) 
print("sep=1")
print(f'Number of support vectors = {len(svm_1.support_vectors_)}')
print(f'reference margin = {1/np.linalg.norm(svm_1.coef_):.3f}')

print()

svm_20 = linear_svm().fit(training_sample_20_X, training_sample_20_y)
print("sep=20")
print(f'Number of support vectors = {len(svm_20.support_vectors_)}')
print(f'reference margin = {1/np.linalg.norm(svm_20.coef_):.3f}')
```

Higher values of class_sep spread out the clusters more, making the classification task easier because the classes are more spread out from each other. There are fewer support vectors for the SVM in the second case, because if the class separation is higher, then more points would be past the reference margin and correctly classified. The support vectors are the points that contribute to the SVM decision boundary: that is, the points that are correctly classified but within the reference margin or are incorrectly classifier. Since the class separation is higher, there are less points that are correctly classifier but within the reference margin. We can see this makes sense as the reference margin for case 2 is 7.667, significantly larger than the reference margin for 0.776 in case 1, which makes sense since the classes are more spread out in case 2 and thus the algorithm was able to find a large $\mu^*$

### Problem 3.2 (Exam Style, except for the programming aspects)

```python
def residual(w_star, x): 
    
    print(x.shape, w_star.shape) 
    
    beta = np.linalg.lstsq(x.T, w_star.T, rcond=None)[0]
    
    summation = np.dot(beta.flatten(), x)
    r_I = np.linalg.norm(w_star - summation)
    return r_I

def show_residual(n,d,sep):
    sample_X, sample_y = samples(n,d,sep)
    svm = linear_svm().fit(sample_X, sample_y)
    support_indices = svm.support_
    x_support_vectors = sample_X[support_indices]

    
    r_t = residual(svm.coef_, sample_X)
    r_s = residual(svm.coef_, x_support_vectors)

    
    print(f'n = {n}')
    print(f'd = {d}')
    print(f'sep = {sep}')
    print(f'nsv ={len(support_indices)}')
    print(f'r_t = {r_t:.2e}')
    print(f'r_s = {r_s:.2e}')
```

```python
show_residual(100, 30, 0.1)
```

The small values of $r_T$ and $r_S$ are not surprising. There are 100 data points and 30 dimensions, which means that it is very likely (not guaranteed, but high probability) that the vectors of these 100 data points span the whole $\mathbb{R}^{30}$. This means that there is a very high chance that $w^*$ is also in the span of these vectors, which means that the least squares calculation for $w^*$ should yield a very small residual.

## Part 4: The RBF Kernel

```python
from sklearn.datasets import make_moons
from types import SimpleNamespace


def moons(n, noise=0.3, seed=1):
    x, y = make_moons(n_samples=n, noise=noise, \
                      random_state=seed, shuffle=True)
    y[y <= 0] = -1
    return SimpleNamespace(x=x, y=y)
```

```python
training_set = moons(n=100)
test_set = moons(n=10000, seed=3)
```

```python
def rbf_svm(c=None, gamma=None):
    return SVC(kernel='rbf', C=c, gamma=gamma, tol=1.e-6,
               decision_function_shape='ovo')
```

```python
from matplotlib import pyplot as plt
from mpl_toolkits.axes_grid1 import Divider, Size
import numpy as np
from sklearn.inspection import DecisionBoundaryDisplay
from sklearn.model_selection import GridSearchCV
```

```python
def data_box(x, pad=0.5):
    x0, x1 = x[:, 0], x[:, 1]
    return [
        np.min(x0) - pad, np.max(x0) + pad,
        np.min(x1) - pad, np.max(x1) + pad
    ]
```

```python
def fixed_axis_figure(box, scale=1., axis='on', h_pads=None, \
                      v_pads=None):
    if h_pads is None:
        h_pads = [0.7, 0.1] if axis == 'on' else [0, 0]
    if v_pads is None:
        v_pads = [0.5, 0.5] if axis == 'on' else [0, 0]
    ax_h_size, ax_v_size = scale * (box[1] - box[0]), \
    scale * (box[3] - box[2])
    fig_h_size, fig_v_size = h_pads[0] + ax_h_size + h_pads[1], \
    v_pads[0] + ax_v_size + v_pads[1]
    fig = plt.figure(figsize=(fig_h_size, fig_v_size))
    horizontal = [Size.Fixed(h_pads[0]), Size.Fixed(ax_h_size), \
                  Size.Fixed(h_pads[1])]
    vertical = [Size.Fixed(v_pads[0]), Size.Fixed(ax_v_size), \
                Size.Fixed(v_pads[1])]
    d = Divider(fig, (0, 0, 1, 1), horizontal, vertical,
                aspect=False)
    ax = fig.add_axes(d.get_position(), \
                      axes_locator=d.new_locator(nx=1, ny=1))
    plt.xlim(box[:2])
    plt.ylim(box[2:])
    plt.axis(axis)
    return fig, ax
```

### Problem 4.1: Displaying an SVM

```python
def plot_classifier(c, gamma, train, test, size=18):
    model = rbf_svm(c, gamma)
    model.fit(train.x, train.y)

    scoretrain =  model.score(train.x, train.y)
    scoretest =  model.score(test.x, test.y)


    x, y = train.x, train.y

    colors = {1: 'tab:blue', -1: 'tab:red'}
    ranges = data_box(x)
    fig, ax = fixed_axis_figure(ranges, scale=1.2, v_pads=[0.5, 0.5])
    for label in (-1, 1):
        idx = y == label
        ax.scatter(x[idx, 0], x[idx, 1], c=colors[label],
                    label='true label {:2d}'.format(label), s=18)
        
    sv = model.support_vectors_
    sv_labels = model.predict(sv)
    for label in (-1, 1):
        idx = sv_labels == label
        ax.scatter(sv[idx, 0], sv[idx, 1], facecolors='none', \
                   edgecolor=colors[label],
                   s=90, label=f'Predicted label {label:2d}')
        
    common_params = {"estimator": model, "X": train.x, "ax": ax}
    DecisionBoundaryDisplay.from_estimator(
        **common_params,
        response_method="predict",
        plot_method="contour",
        levels=np.unique(model.predict(train.x)),
        colors='k', 
        linestyles='-', 
        alpha=1,
    )
    DecisionBoundaryDisplay.from_estimator(
        **common_params,
        response_method="decision_function",
        plot_method="contour",
        levels=[-1, 0, 1],
        colors=["k", "k", "k"],
        linestyles=["--", "-", "--"],
    )
    ax.set_title(
        f"$C_0={c:.3f}, \\gamma={gamma:.3f}$"
        f"$\\text{{, accuracies: train }} {scoretrain:.3f}, \
        \\text{{test }} {scoretest:.3f}$"
    )    
    ax.legend()
    plt.show()
```

```python
plot_classifier(1., 1., training_set, \
                test_set)
```

Qualitatively, gamma is controlling the curvature of the decision boundary, where a higher gamma increases the curvature, and a lower gamma decreases the curvature. It essentially is controlling the amount of overfitting. This makes sense because smaller values of $\gamma$ creates smoother feature space mappings, which would thus make the decision boundaries smoother and reduce curvature.

The SVM overfits the least for $\gamma = 0.125$, which is the smallest value of $\gamma$
